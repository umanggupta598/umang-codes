```python
#importing necessary packages and changing my working directory
import csv
import os
import json
os.chdir(r'C:\Users\umang\OneDrive\Desktop\mainfolder')
```

```python
#to read the data from the desired files
def load_data():
    sales_data = {}
    product_details = {}
    product_descriptions = {}
    

    # Get the CSV file
    csv_file_path = r'sales_data.csv'
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        
        # Read and store the data into sales_data dictionary
        for row in reader:
            # Assuming first column is the key (like product SKU) and second to end is sales data
            sales_data[row[0]] = row[1:]
            
    
    # Process product_details JSON files
    folder_path = r'C:\Users\umang\OneDrive\Desktop\mainfolder\product_details'
    files = os.listdir(folder_path)
    json_files = [file for file in files if file.endswith('.json')]
    for json_file in json_files:
        file_path = os.path.join(folder_path, json_file)
        with open(file_path, 'r') as file:
            data = json.load(file)
            product_details[json_file] = data
            

    # Process product_descriptions text files
    folder_path = r'C:\Users\umang\OneDrive\Desktop\mainfolder\product_descriptions'
    files = os.listdir(folder_path)
    txt_files = [file for file in files if file.endswith('.txt')]
    for txt_file in txt_files:
        file_path = os.path.join(folder_path, txt_file)
        with open(file_path, 'r') as file:
            data = file.read()
            product_descriptions[txt_file] = data
    
    # Return the values to be used outside this function
    return sales_data, product_details, product_descriptions


print(load_data())
        
```

```python
# Function to update sales data
def update_sales_data(sales_data, sales_key, sales_value):
    sales_data[sales_key] = sales_value
    return sales_data

# Function to update product details
def update_product_details(product_details, product_key, product_value):
    product_details[product_key] = product_value
    return product_details

# Function to update product descriptions
def update_product_descriptions(product_descriptions, description_key, description_value):
    product_descriptions[description_key] = description_value
    return product_descriptions
```

```python
# The new update function to take user input and perform validation and calls the update_sales_data, updata_product_details and update_product_description functions
def update():
    # Load the data
    sales_data, product_details, product_descriptions = load_data()

    # Step 1: Collect and validate SKU
    sku = input("Enter the SKU (must be exactly 13 characters): ")
    if len(sku) != 13:
        print("Error: SKU must be exactly 13 characters long.")
        return sales_data, product_details, product_descriptions  # Early exit

    # Step 2: Collect and validate sales data for the last 14 days
    sales_input = input("Enter sales data for the last 14 days (14 whole numbers separated by spaces): ")
    sales_data_list = sales_input.split()
    
    if len(sales_data_list) != 14 or not all(sale.isdigit() for sale in sales_data_list):
        print("Error: Sales data must consist of exactly 14 whole numbers.")
        return sales_data, product_details, product_descriptions  # Early exit

    # Convert sales data to a list of integers
    sales_values = list(map(int, sales_data_list))

    # Step 3: Collect product details (name, brand, model, specifications, price, availability)
    name = input("Enter product name: ")
    brand = input("Enter product brand: ")
    model = input("Enter product model: ")
    specifications = input("Enter product specifications: ")
    price = input("Enter product price: ")
    availability = input("Enter product availability: ")

    # Prepare product details dictionary
    product_info = {
        'name': name,
        'brand': brand,
        'model': model,
        'specifications': specifications,
        'price': price,
        'availability': availability
    }

    # Step 4: Collect product description
    description = input("Enter product description: ")
    if not description:
        print("Error: Product description is required.")
        return sales_data, product_details, product_descriptions  # Early exit

    # Step 5: Update all dictionaries if all inputs are valid
    updated_sales_data = update_sales_data(sales_data, sku, sales_values)
    updated_product_details = update_product_details(product_details,sku, product_info)
    updated_product_descriptions = update_product_descriptions(product_descriptions,sku, description)

    # Print success message
    print("Product successfully registered and updated.")

    # Return updated dictionaries
    return updated_product_details, updated_sales_data, updated_product_descriptions


# Example usage
updated_product_details, updated_sales_data, updated_product_descriptions = update()

# Display the updated data (for checking purposes)
print("\nUpdated Sales Data:")
print(updated_sales_data)

print("\nUpdated Product Details:")
print(updated_product_details)

print("\nUpdated Product Descriptions:")
for file_name, description in updated_product_descriptions.items():
    print(f"\nFile: {file_name}")
    print(description)
```

```python
def ensure_folder_exists(folder_path):
    """Ensure that the folder exists, and create it if it doesn't."""
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

def clear_folder(folder_path):
    """Clear all files from the specified folder. If we are not using this function, it is creating duplicates of the files"""
    if os.path.exists(folder_path):
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)
            if os.path.isfile(file_path):
                os.remove(file_path)

def dump(sales_data, product_details, product_descriptions, main_folder):
    """
    Dump sales_data to a CSV, product_details to separate JSON files, 
    and product_descriptions to separate TXT files.
    """
    # Ensure the main folder exists
    ensure_folder_exists(main_folder)

    # Step 1: Dump sales_data into a CSV file
    csv_file_path = os.path.join(main_folder, 'sales_data.csv')
    with open(csv_file_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for sku, sales in sales_data.items():
            writer.writerow([sku] + sales)

    # Step 2: Dump product_details into separate JSON files
    product_details_folder = os.path.join(main_folder, 'product_details')
    ensure_folder_exists(product_details_folder)
    
    # Clear the folder to prevent duplicates
    clear_folder(product_details_folder)
    
    for sku, details in product_details.items():
        json_file_path = os.path.join(product_details_folder, f"details_{sku}.json")
        with open(json_file_path, 'w') as jsonfile:
            json.dump(details, jsonfile, indent=4)

    # Step 3: Dump product_descriptions into separate TXT files
    product_descriptions_folder = os.path.join(main_folder, 'product_descriptions')
    ensure_folder_exists(product_descriptions_folder)
    
    # Clear the folder to prevent duplicates
    clear_folder(product_descriptions_folder)
    
    for sku, description in product_descriptions.items():
        txt_file_path = os.path.join(product_descriptions_folder, f"description_{sku}")
        with open(txt_file_path, 'w') as txtfile:
            txtfile.write(description)

    print("Data has been successfully dumped to files.")
```

```python
# Now calling all functions here separately, load_data, update and dump
main_folder = r'C:\Users\umang\OneDrive\Desktop\mainfolder'

# Load data (assuming load_data function exists and works as expected)
sales_data, product_details, product_descriptions = load_data()

# Update the dictionaries
updated_product_details, updated_sales_data, updated_product_descriptions = update()

# Now you can call the dump function independently
dump(updated_sales_data, updated_product_details, updated_product_descriptions, main_folder)
```
